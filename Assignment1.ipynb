```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
file_path = r"C:\Users\User\Downloads\Basic stats - 1 (1)\Basic stats - 1\sales_data_with_discounts.csv"

data = pd.read_csv(file_path)

# Display the first few rows
print("Dataset Preview:")
print(data.head())
print("\nDataset Info:")
print(data.info())
print("\nSummary Statistics:")
print(data.describe())

# Step 2: Descriptive Analytics for Numerical Columns
print("\nDescriptive Analytics for Numerical Columns:")
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns
print("Numerical Columns:", numerical_cols)

# Compute basic statistics
for col in numerical_cols:
    print(f"\nStatistics for {col}:")
    print(f"Mean: {data[col].mean():.2f}")
    print(f"Median: {data[col].median():.2f}")
    print(f"Mode: {data[col].mode()[0]:.2f}")
    print(f"Standard Deviation: {data[col].std():.2f}")

# Step 3: Data Visualization

# Histograms
print("\nPlotting Histograms...")
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True, bins=30, color='blue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Boxplots
print("\nPlotting Boxplots...")
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data[col], color='green')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()

# Bar Chart Analysis for Categorical Columns
categorical_cols = data.select_dtypes(include=['object', 'category']).columns
print("\nCategorical Columns:", categorical_cols)

print("\nPlotting Bar Charts...")
for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    data[col].value_counts().plot(kind='bar', color='orange')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

# Step 4: Standardization of Numerical Variables
print("\nStandardizing Numerical Variables...")
standardized_data = data[numerical_cols].apply(lambda x: (x - x.mean()) / x.std())

# Compare before and after distributions
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.kdeplot(data[col], label='Original', color='blue')
    sns.kdeplot(standardized_data[col], label='Standardized', color='red')
    plt.title(f'Distribution Before and After Standardization: {col}')
    plt.legend()
    plt.show()

# Step 5: Conversion of Categorical Data into Dummy Variables
print("\nConverting Categorical Data into Dummy Variables...")
encoded_data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
print("Transformed Dataset with Dummy Variables:")
print(encoded_data.head())

# Step 6: Conclusion
print("\nData Preprocessing Complete!")
print("Key findings and transformations:")
print("- Descriptive statistics calculated for numerical columns.")
print("- Histograms and boxplots visualized data distributions and outliers.")
print("- Numerical variables standardized for uniformity.")
print("- Categorical variables converted to dummy variables for analysis and modeling.")
```
