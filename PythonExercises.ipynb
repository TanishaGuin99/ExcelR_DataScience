```python
# Exercise 1: Prime Numbers
# A program to check if a given number is prime.
def is_prime(number):
    """
    Function to check whether a number is prime.
    Arguments:
    number -- The number to check

    Returns:
    True if number is prime, otherwise False.
    """
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

# Test the function
num = int(input("Enter a number to check if it is prime: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")


# Exercise 2: Product of Random Numbers
import random

def product_game():
    """
    Function to generate two random numbers, ask user for their product, and check the answer.
    """
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)

    print(f"What is the product of {num1} and {num2}?")
    user_answer = int(input("Your answer: "))

    if user_answer == num1 * num2:
        print("Correct!")
    else:
        print(f"Incorrect. The correct answer is {num1 * num2}.")

# Run the game
product_game()


# Exercise 3: Squares of Even/Odd Numbers
# Printing the squares of even numbers between 100 and 200.
# Change `even` to `odd` in the condition to print squares of odd numbers instead.
def print_squares():
    """
    Function to print squares of even numbers between 100 and 200.
    """
    print("Squares of even numbers between 100 and 200:")
    for number in range(100, 201):
        if number % 2 == 0:  # Change to `if number % 2 != 0:` for odd numbers
            print(f"{number}^2 = {number ** 2}")

# Run the function
print_squares()


# Exercise 4: Word Counter
def word_counter(input_text):
    """
    Function to count the occurrences of each word in the given text.
    Arguments:
    input_text -- The text to analyze

    Returns:
    Dictionary with words as keys and their counts as values.
    """
    words = input_text.split()
    word_counts = {}
    for word in words:
        word_counts[word] = word_counts.get(word, 0) + 1
    return word_counts

# Example input
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
word_counts = word_counter(input_text)

# Print the word counts
print("Word counts:")
for word, count in word_counts.items():
    print(f"'{word}': {count}")


# Exercise 5: Check for Palindrome
def is_palindrome(input_string):
    """
    Function to check if a string is a palindrome.
    Arguments:
    input_string -- The string to check

    Returns:
    True if the string is a palindrome, otherwise False.
    """
    # Remove spaces, punctuation, and convert to lowercase
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())
    return cleaned_string == cleaned_string[::-1]

# Test the function
input_string = input("Enter a string to check if it is a palindrome: ")
if is_palindrome(input_string):
    print(f"'{input_string}' is a palindrome.")
else:
    print(f"'{input_string}' is not a palindrome.")
```
